```python
region_code = pd.read_csv('Rate.csv')
region_code.head()
```

```python
lpd_2020 = pd.read_csv('LPD2020_target_final.csv')
lpd_2020.head()
```

```python
area = ['삼각지역', '군자역', '회기역', '용산역']
cols = ['TOT_REG_CD','AREA_NM', 'ADM_CD', 'RATE']
region_code_edit = region_code.loc[region_code['AREA_NM'].isin(area), cols]
region_code_edit
```

```python
merged_df = pd.merge(lpd_2020, region_code_edit, left_on='행정동코드', right_on='ADM code', how='inner')
merged_df.head()
```
